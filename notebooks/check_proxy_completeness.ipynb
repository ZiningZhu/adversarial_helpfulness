## Proxy evaluation results

```python
import numpy as np 
from pathlib import Path 
import scipy
import pandas as pd
```

```python
def check_completeness():

    evaluators = ["mixtral-8x7B", "vicuna-33B"]
    explainers = ["gpt4", "chat", "claude"]
    tasks = {"ecqa": ["secondbest"], "nli": ["entail_to_neutral", "contra_to_neutral"]}

    report = {"dataset": [], "task": [], "explainer": [], "evaluator": [], "has_nle": [], "has_score": []}
    for ds in tasks:
        for t in tasks[ds]:
            for exp in explainers:
                for ev in evaluators:
                    nle_pth = Path(f"../data/{ds}/{t}/{exp}/with_nle.csv")
                    report["has_nle"].append(nle_pth.exists())
                    fn_pth = Path(f"../data/{ds}/{t}/{exp}/with_nle_scored_by_{ev}.csv")
                    report["has_score"].append(fn_pth.exists())
                    report["dataset"].append(ds)
                    report["task"].append(t)
                    report["explainer"].append(exp)
                    report["evaluator"].append(ev)

    return pd.DataFrame(report)

check_completeness()
```
